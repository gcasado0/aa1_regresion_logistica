Para determinar el número óptimo de regiones (k) en las que agrupar las ubicaciones del dataset, se utilizó el "Método del Codo". Este método consiste en utilizar el algoritmo "K-Means" para diferentes valores de k y calcular la inercia (suma de las distancias al cuadrado dentro de cada clúster) para cada k.

```python
# Escalamos los datos para aplicar k-means
X = df[['Latitud', 'Longitud']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertias = []
max_k = 10

# Calculamos la inercia para cada k
for k in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)


plt.figure(figsize=(15,4))
plt.plot(range(1, max_k + 1), inertias, marker='o', linestyle='--')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.grid(True)
plt.xticks(range(1, max_k + 1))
plt.show()
```

Este método muestra un quiebre pronunciado en k = 4 debido a que se ve una caída fuerte hasta este valor y luego el descenso es mucho más leve. 

Esto indica que agregar más clusters más allá de 4 no aporta una mejora significativa en la compactación de los grupos. 

Por esta razón, concluimos que el valor óptimo de clusters para dividir las localizaciones es 4 y podrían interpretarse como Norte, Sur, Este y Oeste.

```python
# Utiliza el algoritmo de k-means con 4 clusters (regiones)
kmeans = KMeans(n_clusters = 4, random_state=42, n_init = 10)
df['Regiones'] = kmeans.fit_predict(X_scaled)
```

```python
# Crear un diccionario para mapear las etiquetas numéricas
etiqueta = {0: 'Sur', 1: 'Norte', 2: 'Este', 3: 'Oeste'}
df['Regiones'] = df['Regiones'].map(etiqueta)

df.head()
```

#### Análisis y decisión sobre datos faltantes

```python
# Observamos las variables que poseen valores faltantes en sus datos
df.isnull().sum()
```

```python
def calcular_porcentaje(faltante, total):
    '''
    Evita la división por cero al calcular el porcentaje.
    '''
    if total == 0:
        return 0
    
    porcentaje = (faltante / total) * 100
    return porcentaje 

def imprimir_porcentaje(porcentaje, columna):
    '''
    Imprime los resultados del porcentaje por cada columna.
    '''
    print(f"{columna}: {porcentaje:.2f}%")
    return

def datos_faltantes(columna):
    '''
    Calcula el porcentaje de valores faltantes de cada columna
    '''
    faltante = df[columna].isnull().sum()
    total = len(df)
    porcentaje = calcular_porcentaje(faltante, total)
    imprimir_porcentaje(porcentaje, columna)
    return 

print("Porcentaje de los valores faltantes de todas las columnas:")
for columna in df.columns:
    datos_faltantes(columna)
```

Podemos observar que las columnas que poseen un mayor valor de faltantes son los siguientes:
- Evaporation: 43.17%

- Cloud9am: 38.42%

El resto de valores tiene un porcentaje menor al 11%, por lo tanto, se considera adecuado imputar estos datos utilizando un método apropiado después de dividir el dataset en conjuntos de entrenamiento y prueba. 

Las columnas con un alto porcentaje de datos faltantes (30-50%) serán eliminadas del dataset. Imputar una gran cantidad de valores puede generar ruido y afectar al rendimiento predictivo del modelo.

```python
# funciones para imputar datos faltantes usando la mediana o la moda

def imput_median(columna):
    mediana= train_df[columna].median()
    print(f"Mediana de {columna}: {round(mediana,2)}")
    print(f"Total de casos imputados con la mediana en train: { train_df[columna].isnull().sum()} ")
    print(f"Total de casos imputados con la mediana en test: {test_df[columna].isnull().sum()} ")
    return

def imput_mode(columna):
    moda = train_df[columna].mode()[0]
    print(f"Moda de {columna}: {moda}")
    print(f"Total de casos imputados con la moda en train: { train_df[columna].isnull().sum()} ")
    print(f"Total de casos imputados con la moda en test: {test_df[columna].isnull().sum()} ")
    return
```

```python
import math
def boxplots_target(X, y):
  
    feature_names=X.columns.values
    target_name= y.name

    data = pd.concat([pd.DataFrame(y, columns=[target_name]), pd.DataFrame(X, columns=feature_names)], axis=1)

    num_features = X.shape[1]
    num_rows = math.ceil(num_features / 3)

    fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5 * num_rows), squeeze=False)
    fig.subplots_adjust(hspace=0.5)

    # Se recorre cada feature para generar su diagrama de caja respecto de la salida
    for i, feature in enumerate(data.columns[1:]):
        row = i // 3
        col = i % 3
        sns.boxplot(x=target_name, y=feature, data=data, ax=axes[row, col])
        axes[row, col].set_title(f'Diagrama de caja de {feature}')
        axes[row, col].set_xlabel('Target')
        axes[row, col].set_ylabel(feature)

    # Eliminar subplots vacíos si es necesario
    if num_features % 3 != 0:
        for i in range(num_features % 3, 3):
            fig.delaxes(axes[num_rows - 1, i])

    plt.show()
```

### Modelo de regresión logística sin balancear los datos

```python
# creo variable para guardar los resultados de los modelos
resultados_modelos = pd.DataFrame()
```

```python
# columnas modelo, clase, precision, recall, f1-score, support
# crear funcion que agregue los resultados del classification report al dataframe resultados_modelos
def agregar_resultados_modelo(modelo_nombre, y_true, y_pred):
    global resultados_modelos
    report = classification_report(y_true, y_pred, output_dict=True)
    rows = []
    for clase, metrics in report.items():
        if clase not in ['accuracy', 'macro avg', 'weighted avg']:
            row = {
                'Modelo': modelo_nombre,
                'Clase': clase,
                'Precision': metrics.get('precision', None),
                'Recall': metrics.get('recall', None),
                'F1-Score': metrics.get('f1-score', None),
                'Support': metrics.get('support', None)
            }
            rows.append(row)
    if rows:
        resultados_modelos = pd.concat([resultados_modelos, pd.DataFrame(rows)], ignore_index=True)
    return
```

```python
# Entrena un modelo de regresión logística sin balancear los datos
logistic_model = LogisticRegression(random_state=42, max_iter=150)
logistic_model.fit(X_train, y_train)
y_pred_s = logistic_model.predict(X_test)

print("Resultados con regresión logística sin balanceo\n")
print("\n---")
print("Classification Report:\n", classification_report(y_test, y_pred_s))
print("Balanced accuracy: ", round(balanced_accuracy_score(y_test, y_pred_s),2))

# agregar resultados del modelo sin balanceo
agregar_resultados_modelo('Regresión Logística Sin Balanceo', y_test, y_pred_s)
```

- Ambos modelos con balanceo y sin balanceo tienen un valor AUC practicamente igual de 0.86.

- Una forma de calcular el mejor umbral es calcular la distancia desde cada punto de la curva al punto (0,1) y encontrar el mínimo.

```python
# Calcular la distancia desde cada punto al (0, 1)
# √ (FPR - 0)2 + (TPR - 1)2
distances = np.sqrt(fpr1**2 + (tpr1-1)**2)

# Encontrar el índice (ix) de la distancia mínima
ix = np.argmin(distances)

# Obtener el umbral óptimo
optimal_threshold_geom1 = thresholds1[ix]

print(f'Umbral óptimo (Geométrico) Modelo balanceado: {optimal_threshold_geom1:.4f}')
print(f'En este umbral:')
print(f'  Tasa de Verdaderos Positivos (TPR): {tpr1[ix]:.4f}')
print(f'  Tasa de Falsos Positivos (FPR): {fpr1[ix]:.4f}')
```

```python
# Calcular la distancia desde cada punto al (0, 1)
distances = np.sqrt(fpr2**2 + (tpr2-1)**2)

# Encontrar el índice (ix) de la distancia mínima
ix = np.argmin(distances)

# Obtener el umbral óptimo
optimal_threshold_geom2 = thresholds2[ix]

print(f'Umbral óptimo (Geométrico) Modelo sin balancear: {optimal_threshold_geom2:.4f}')
print(f'En este umbral:')
print(f'  Tasa de Verdaderos Positivos (TPR): {tpr1[ix]:.4f}')
print(f'  Tasa de Falsos Positivos (FPR): {fpr1[ix]:.4f}')
```

- Si el modelo es el balanceado el umbral 0.5 se aproxima al óptimo: 0.4755. Pero con el modelo sin balancear el umbral óptimo sería: 0.2284.

```python
# obtener probabilidades y aplicar umbral óptimo del modelo balanceado
y_pred1_thresh = (y_probs1 >= optimal_threshold_geom1).astype(int)

# métricas con el umbral aplicado
print(f"Classification report (modelo balanceado, umbral óptimo): {round(optimal_threshold_geom1,2)}")
print(classification_report(y_test, y_pred1_thresh))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred1_thresh))
print("Balanced accuracy: ", round(balanced_accuracy_score(y_test, y_pred1_thresh),2))
```

### Búsqueda en grilla (Grid Search):

```python
def grid_search(parameters, score, X_train, y_train, X_test, y_test):
    '''
    Realiza una búsqueda en grilla (Grid Search) para encontrar los mejores
    hiperparámetros para un modelo LogisticRegression().
    '''
    grid_search_cv = GridSearchCV(estimator = LogisticRegression(), param_grid = parameters, cv = 5, scoring = score, n_jobs = -1)
    start = time.time()
    grid_search_cv.fit(X_train, y_train)
    end = time.time()
    grid_search_time = end - start
    y_pred_grid = grid_search_cv.predict(X_test)

    best_grid = grid_search_cv.best_estimator_
    report_grid = classification_report(y_test, y_pred_grid)
    print(grid_search_cv.best_params_)

    # Agregar resultados del modelo con Grid Search 
    agregar_resultados_modelo('Regresión Logística con Grid Search', y_test, y_pred_grid)

    return grid_search_time, best_grid, report_grid
```

### Búsqueda aleatoria (Random Search):

```python
def random_search(parameters, score, X_train, y_train, X_test, y_test):
    '''
    Realiza una búsqueda aleatoria (Random Search) para encontrar los mejores
    hiperparámetros para un modelo LogisticRegression().
    '''
    random_search_cv = RandomizedSearchCV(
    estimator=LogisticRegression(),
    param_distributions=parameters,
    n_iter=20,                      # num de combinaciones a probar
    scoring= score,
    cv=5,     
    random_state=42,
    n_jobs=-1)

    start = time.time()
    # Entrenamos al modelo
    random_search_cv.fit(X_train, y_train)
    end = time.time()
    random_search_time = end - start
    
    y_pred_random = random_search_cv.predict(X_test)
    best_random = random_search_cv.best_estimator_
    report_random = classification_report(y_test, y_pred_random)
    print(random_search_cv.best_params_)

    # Agregar resultados del modelo con Random Search
    agregar_resultados_modelo('Regresión Logística con Random Search', y_test, y_pred_random)

    return random_search_time, best_random, report_random
```

### Optimización bayesiana:

```python
def objective(trial, X, y):
    class_weight = trial.suggest_categorical('class_weight', ['balanced', None])
    C = trial.suggest_float('C', 0.1, 100, log=True)
    solver = 'newton-cg'
    max_iter = trial.suggest_int('max_iter', 100, 1000)

    params = {
        'class_weight': class_weight,
        'C': C,
        'solver': solver,
        'max_iter': max_iter
    }

    model = LogisticRegression(**params)
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X, y, cv=kfold, scoring='f1', n_jobs=-1)

    return np.mean(scores)
```

```python
def optuna_search(objective, n_trials, X_train, y_train, X_test, y_test):
    """
    Realiza una optimización bayesiana para encontrar los mejores
    hiperparámetros para un modelo LogisticRegression().
    """
    study = optuna.create_study(direction='maximize')
    start = time.time()

    # "Envolvemos" objective para que X_train e y_train ya estén "incluidos"
    objective_con_datos = lambda trial: objective(trial, X_train, y_train)
    study.optimize(objective_con_datos, n_trials=n_trials)
    
    end = time.time()
    optuna_time = end - start
    best_optuna = study.best_params
    best_optuna['solver'] = 'newton-cg'
    final_model = LogisticRegression(**best_optuna, random_state=42, n_jobs=-1)
    final_model.fit(X_train, y_train)
    y_pred_optuna = final_model.predict(X_test)
    report_optuna = classification_report(y_test, y_pred_optuna)

    # Agregar resultados del modelo con Optuna
    agregar_resultados_modelo('Regresión Logística con Optuna', y_test, y_pred_optuna)

    return optuna_time, best_optuna, report_optuna
```

### Sin optuna

```python
class NeuralNetwork:
    def __init__(self, epochs, batch_size, learning_rate):
        #inicializo algunos parámetros como épocas, batch_size, learning rate        
        self.epochs = epochs
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.model = None

    def build_model(self, input_shape):
        model = tf.keras.models.Sequential([
            # Capa de entrada
            tf.keras.layers.Dense(64, activation='relu', input_shape=(input_shape,)),
            tf.keras.layers.Dropout(0.3),
            # Capa oculta
            tf.keras.layers.Dense(32, activation='relu'),
            tf.keras.layers.Dropout(0.3),
            # Capa de salida
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])

        optimizer = tf.keras.optimizers.Adam(learning_rate=self.learning_rate)
        # threshold= 0.2: Convierte la probabilidad >= 0.2 a la clase positiva (1)
        # average='macro': Devuelve un solo valor promedio
        metrica_f1 = tf.keras.metrics.F1Score(threshold= 0.2, average='macro', name='f1_score')
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=[metrica_f1])
        
        self.model = model
        model.summary()

    def train(self, X_train, y_train, X_valid, y_valid):
        early_stopping = EarlyStopping(
            monitor='val_f1_score',    # Monitorea la métrica f1-score sobre el conjunto de validación
            patience=10,               # Nro. de epochs sin mejora antes de parar
            restore_best_weights=True  # Se queda con el mejor modelo
        )
        
        history = self.model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=self.epochs, batch_size=self.batch_size, callbacks=[early_stopping])
        return history.history['loss'], history.history['val_f1_score']

    def evaluate(self, X_test, y_test):
        ### evalúo en test
        loss, f1 = self.model.evaluate(X_test, y_test)
        print(f"test F1-Score: {f1:.4f}")

    def predict(self, X_new):
        ### predicciones
        predictions = self.model.predict(X_new)
        return predictions
```

```python
# Parámetros de entrenamiento
nn = NeuralNetwork(epochs=150, batch_size=256, learning_rate=0.001)
```

Implementamos una red neuronal sin optuna con dos capas ocultas (la primer capa con 64 neuronas y la segunda capa con 32 neuronas) con una función de activación ReLu para evitar problemas con los cálculos de los gradientes saturados y una capa de salida con una función de activación sigmoide debido que genera una salida en un rango de valores entre 0 y 1, lo cual lo hace ideal para problemas de clasificación (donde 0 es "No Lluvia" y 1 es "Sí Lluvia"). 

Regularización:
- En las capas ocultas utilizamos una regularización de dropout con una tasa de 0.3 para evitar el sobreajuste (overfitting) del modelo de los datos de entrenamiento. 

Evaluación del rendimiento:
- Utilizamos una detección temprana para monitorear el rendimiento del modelo con la métrica de f1-score sobre el conjunto de validación con una paciencia de 10 épocas sin mejora antes de detener el entrenamiento.
- Dado que los datos están desbalanceados, se utilizó un umbral de decisión de 0.2 para convertir la probabilidad de salida en la clase positiva ("Sí Lluvia") si es mayor o igual a ese valor.

Con los datos obtenidos con optuna lo aplicamos a la clase "NeuralNetwork2 para generar el modelo de redes neuronales:
- Cantidad de capas: 1
- Tasa de aprendizaje: 0.001
- Cantidad de neuronas: 106

Funciones de activación:
- Capas ocultas: ReLu.
- Capa de salida: Sigmoide.
- Utilizamos una función de activación sigmoide debido que genera una salida en un rango de valores entre 0 y 1, lo cual lo hace ideal para problemas de clasificación (donde 0 es "No Lluvia" y 1 es "Sí Lluvia"). Al utilizar ReLu junto la función sigmoide solucionamos el problema de los gradientes saturados generados por la sigmoide. 

Regularización:
- En las capas ocultas utilizamos una regularización de dropout con una tasa de 0.3 para evitar el sobreajuste (overfitting) del modelo de los datos de entrenamiento. 

Evaluación del rendimiento:
- Utilizamos una detección temprana para monitorear el rendimiento del modelo con la métrica de f1-score sobre el conjunto de validación con una paciencia de 10 épocas sin mejora antes de detener el entrenamiento.
- Dado que los datos están desbalanceados, se utilizó un umbral de decisión de 0.2 para convertir la probabilidad de salida en la clase positiva ("Sí Lluvia") si es mayor o igual a ese valor. 

```python
class NeuralNetwork2:
    def __init__(self, epochs, batch_size, learning_rate):        
        self.epochs = epochs
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.model = None

    def build_model(self, input_shape):
        model = tf.keras.models.Sequential([
            # Capa de entrada
            tf.keras.layers.Dense(106, activation='relu', input_shape=(input_shape,), kernel_regularizer=l2(0.0001)),
            tf.keras.layers.Dropout(0.3),

            # Capa de salida
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])

        optimizer = tf.keras.optimizers.Adam(learning_rate=self.learning_rate)
        metrica_f1 = tf.keras.metrics.F1Score(threshold=0.2, average='macro', name='f1_score')
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=[metrica_f1])
        
        self.model = model
        model.summary()

    def train(self, X_train, y_train, X_valid, y_valid):
        early_stopping = EarlyStopping(
            monitor='val_f1_score', # Monitorea la métrica de validación
            patience=10,             # Nro. de epochs sin mejora antes de parar
            restore_best_weights=True # Se queda con el mejor modelo
        )
        
        history=self.model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=self.epochs, batch_size=self.batch_size, callbacks=[early_stopping])
        return history.history['loss'], history.history['val_f1_score']

    def evaluate(self, X_test, y_test):
        ### evalúo en test
        loss, f1 = self.model.evaluate(X_test, y_test)
        print(f"test F1-Score: {f1:.4f}")

    def predict(self, X_new):
        ### predicciones
        predictions = self.model.predict(X_new)
        return predictions
```

```python
# Definir el transformador personalizado para la fecha

def extract_month(X):
    """Extrae el mes de una columna de fechas."""
    
    # X llega como un DataFrame de (n_samples, 1)
    # Seleccionamos esa única columna con .iloc[:, 0] para obtener una pd.Series
    X_col = X.iloc[:, 0]

    # Convertir a datetime. 'coerce' convierte errores (o NaNs) en NaT.
    dates = pd.to_datetime(X_col, errors='coerce')
    
    # Extraer el mes. NaT (Not a Time) se convertirá en NaN.
    months = dates.dt.month
    
    # Devolverlo como (n_samples, 1) para sklearn
    return months.values.reshape(-1, 1)

def get_month_feature_names(transformer, input_features):
    return ['Month']

# Creamos el transformador basado en la función
month_transformer = FunctionTransformer(extract_month, validate=False, 
                                        feature_names_out=get_month_feature_names)

# 2. Creamos el pipeline para la fecha (tratar el mes como categórico)
date_pipeline = Pipeline(steps=[
    ('month_extractor', month_transformer),
    ('imputer_month', SimpleImputer(strategy='most_frequent')), 
    ('ohe_month', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Definir el transformador personalizado para la Amplitud Térmica
def calculate_thermal_amplitude(X):
    """Calcula la Amplitud Térmica: MaxTemp - MinTemp."""
    
    # 1. Asegura que sean numéricos (como ya lo tenías)
    # X viene como un array (N, 2) con [MinTemp, MaxTemp]
    max_temp = pd.to_numeric(X[:, 1], errors='coerce') 
    min_temp = pd.to_numeric(X[:, 0], errors='coerce')

    # 2. Calcula la amplitud térmica (el resultado es un numpy array)
    thermal_amplitude = (max_temp - min_temp)
    
    # 3. Haz reshape DIRECTAMENTE sobre el array de NumPy para que 
    # tenga la forma (n_samples, 1) esperada por el ColumnTransformer.
    return thermal_amplitude.reshape(-1, 1)

def get_amplitude_feature_names(transformer, input_features):
    return ['AmplitudTermica']

# Creamos el transformador basado en la función
thermal_amplitude_transformer = FunctionTransformer(calculate_thermal_amplitude, validate=False,
        feature_names_out=get_amplitude_feature_names)

mapping_dict = {'Albury': 'Sur',
 'BadgerysCreek': 'Este',
 'Cobar': 'Este',
 'CoffsHarbour': 'Este',
 'Moree': 'Este',
 'Newcastle': 'Este',
 'NorahHead': 'Este',
 'NorfolkIsland': 'Este',
 'Penrith': 'Este',
 'Richmond': 'Sur',
 'Sydney': 'Este',
 'SydneyAirport': 'Este',
 'WaggaWagga': 'Sur',
 'Williamtown': 'Este',
 'Wollongong': 'Este',
 'Canberra': 'Sur',
 'Tuggeranong': 'Sur',
 'MountGinini': 'Sur',
 'Ballarat': 'Sur',
 'Bendigo': 'Sur',
 'Sale': 'Sur',
 'MelbourneAirport': 'Sur',
 'Melbourne': 'Sur',
 'Mildura': 'Sur',
 'Nhil': 'Sur',
 'Portland': 'Sur',
 'Watsonia': 'Sur',
 'Dartmoor': 'Sur',
 'Brisbane': 'Este',
 'Cairns': 'Norte',
 'GoldCoast': 'Este',
 'Townsville': 'Norte',
 'Adelaide': 'Sur',
 'MountGambier': 'Sur',
 'Nuriootpa': 'Sur',
 'Woomera': 'Sur',
 'Albany': 'Oeste',
 'Witchcliffe': 'Oeste',
 'PearceRAAF': 'Norte',
 'PerthAirport': 'Oeste',
 'Perth': 'Oeste',
 'SalmonGums': 'Oeste',
 'Walpole': 'Oeste',
 'Hobart': 'Sur',
 'Launceston': 'Sur',
 'AliceSprings': 'Norte',
 'Darwin': 'Norte',
 'Katherine': 'Norte',
 'Uluru': 'Norte'}

def map_location_to_region(X, mapping):
    """Mapea la columna Location a Regiones usando un diccionario."""
    
    # X llega como DataFrame (n_samples, 1), lo convertimos a pd.Series
    X_col = X.iloc[:, 0]
    
    # .map() aplica el diccionario.
    # Si una 'Location' no está en el 'mapping', pondrá NaN.
    regions = X_col.map(mapping)
    
    # Devolver como (n_samples, 1) para sklearn
    return regions.values.reshape(-1, 1)

def get_region_feature_names(transformer, input_features):
    return ['Regiones']

# El FunctionTransformer que usa la función map_location_to_region y el diccionario mapping_dict
region_mapper = FunctionTransformer(
    map_location_to_region,
    kw_args={'mapping': mapping_dict},  # <-- Pasa el diccionario a la función
    validate=False,
    feature_names_out=get_region_feature_names # Nombra la salida
)

# Definición de variables y modelo 

# Columnas a imputar con la mediana
variables_numericas = [
    "MinTemp",
    "MaxTemp",
    "WindGustSpeed",
    "WindSpeed9am",
    "WindSpeed3pm",
    "Humidity9am",
    "Humidity3pm",
    "Pressure9am",
    "Pressure3pm",
    "Temp9am",
    "Temp3pm",
    "Rainfall"
]

variables_categoricas = [
    "WindGustDir",
    "WindDir3pm",
    "WindDir9am"
]

variable_fecha = ["Date"]
variable_raintoday = ["RainToday"]
variable_location = ["Location"]

# Columnas necesarias para el cálculo de la Amplitud Térmica
variables_amplitud_termica = ["MinTemp", "MaxTemp"]

# 3. Modificar el ColumnTransformer

# Definimos el orden: 'No' será 0, 'Yes' será 1
raintoday_categories = [['No', 'Yes']]

# Pipeline para RainToday
raintoday_pipeline = Pipeline(steps=[
    ("imputer_mode", SimpleImputer(strategy='most_frequent')),
    ("encoder", OrdinalEncoder(categories=raintoday_categories, dtype=int))
])

# Pipeline para el cálculo de la Amplitud Térmica
amplitude_pipeline = Pipeline(steps=[    
    # Usamos un SimpleImputer con 'median' para ser coherentes con tu imputación numérica.
    ("imputer_minmax", SimpleImputer(strategy='median')), 
    ("amplitude_calc", thermal_amplitude_transformer),
    ("scale", RobustScaler()) # Escalar la nueva característica
])

# Pipeline completo para 'Location'
location_pipeline = Pipeline(steps=[
    ('map_to_region', region_mapper),
    ('imputer_region', SimpleImputer(strategy='most_frequent')), # Rellena NaNs
    ('ohe_region', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Pipeline para preprocesar variables Numericas
numeric_transformer = Pipeline(steps=[
    ("imputer_median", SimpleImputer(strategy='median')),
    ("scale", RobustScaler())
])

# Pipeline para preprocesar variables categóricas
categorical_transformer = Pipeline(steps=[
    ("imputer_mode", SimpleImputer(strategy='most_frequent')),
    ("one_hot_encoder", OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Combinar los transformadores en un ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    # 1. El nuevo transformador para la Amplitud Térmica
    ("amplitude", amplitude_pipeline, variables_amplitud_termica),
    # 2. El transformador numérico para las variables numéricas
    ("numeric", numeric_transformer, variables_numericas),
    # 3. El transformador categórico
    ("categorical", categorical_transformer, variables_categoricas),
    # 4. El transformador para la fecha
    ("date_processing", date_pipeline, variable_fecha),
    # 5. El transformador para RainToday
    ("raintoday_binary", raintoday_pipeline, variable_raintoday),
    # 6. Mapeo de Location a Regiones
    ("location_processing", location_pipeline, variable_location)
])

# Modelo (óptimo según tu comentario)
modelo_optimo = LogisticRegression(class_weight='balanced', C=2, max_iter=500, solver='newton-cg')

# Pipeline final que une el preprocesamiento y el modelo
pipe = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", modelo_optimo)
])
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin

class KerasPretrainedClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, keras_model, threshold=0.2):
        """
        Wrapper para usar un modelo de Keras YA entrenado dentro de un Pipeline de sklearn.
        """
        self.keras_model = keras_model
        self.threshold = threshold

    def fit(self, X, y=None):
        # ¡Importante! No hacemos nada aquí.
        # Como el modelo ya está entrenado, este método es un "dummy"
        # para satisfacer a scikit-learn.
        return self

    def predict(self, X):
        # 1. Obtenemos las probabilidades del modelo Keras
        probs = self.keras_model.predict(X)
        
        # 2. Aplicamos el umbral para devolver clases (0 o 1)
        # Flatten() convierte [[0.1], [0.9]] a [0.1, 0.9]
        predictions = (probs > self.threshold).astype(int).flatten()
        return predictions

    def predict_proba(self, X):
        # Método necesario si quieres calcular curvas ROC o AUC después
        probs = self.keras_model.predict(X)
        # Scikit-learn espera dos columnas para clasificación binaria: 
        # [prob_clase_0, prob_clase_1]
        return np.hstack([1 - probs, probs])

# --- Usarlo en el Pipeline ---

# Envuelves tu modelo existente
modelo_final_wrapper = KerasPretrainedClassifier(nn2, threshold=0.2)
```
